```python
import pandas as pd
```

```python
df = pd.read_csv('chicago.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
columns_to_drop = ['Unnamed: 0', 'unique_key', 'case_number', 'block', 'iucr', 'domestic', 'beat', 'district', 'ward', 'community_area', 'fbi_code', 'x_coordinate', 'y_coordinate', 'year', 'updated_on', 'location']
new_df = df.drop(columns_to_drop, axis=1)
```

```python
new_df.head()
```

```python
new_df.isnull().sum()
```

```python
# Take the rows where location_description is not null
new_df = new_df[new_df['location_description'].notna()]
new_df = new_df[new_df['latitude'].notna()]
new_df = new_df[new_df['longitude'].notna()]
```

```python
new_df.isnull().sum()
```

```python
new_df.head()
```

```python
# Save trimmed dataset
new_df.to_csv('chicago_trimmed.csv', index=False)
```
